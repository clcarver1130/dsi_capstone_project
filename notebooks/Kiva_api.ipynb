```python
# Import libraries
import requests
import pandas as pd
import numpy as np
```

```python
url = 'http://api.kivaws.org/v1/loans/search.json'
params = {'partners': np.linspace(60000, 65000, num=100)}
r = requests.get(url, params=params)
files = r.json()
```

```python
df = pd.DataFrame(files['loans'])
df.head()
```
